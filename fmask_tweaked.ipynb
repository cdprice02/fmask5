```python
import sys
from pathlib import Path
FMASK_PATH = Path("fmask")
sys.path.append(str(FMASK_PATH))
sys.path.append(str(FMASK_PATH.joinpath("src")))

import os
import shutil
import numpy as np
import utils
```

```python
from main.fmask import fmask_physical

HI_scenes_dir = FMASK_PATH.parent.joinpath("cloud-data/HI")
HI_scenes = [os.path.join(HI_scenes_dir, f) for f in os.listdir(HI_scenes_dir)]

dcloud = 0
dshadow = 5
dsnow = 0
out_dir = FMASK_PATH.parent.joinpath(f"out/HI_{dcloud}_{dshadow}_{dsnow}")
if not os.path.exists(out_dir):
    os.makedirs(out_dir)

for scene_path in HI_scenes:
    fmask_physical(
        scene_path,
        dcloud=dcloud,
        dshadow=dshadow,
        dsnow=dsnow,
        destination=out_dir,
        skip=True,
        metadata=False,
        display_fmask=False,
        display_image=False,
        print_summary=True
    )
    
out = shutil.make_archive(out_dir, "gztar", out_dir)
print(f"All scenes processed and saved to {out}.")
```
